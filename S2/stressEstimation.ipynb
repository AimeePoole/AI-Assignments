```python
import tensorflow as tf
from tensorflow import keras

import matplotlib.pyplot as plt
import tensorflow.keras.models 
import tensorflow.keras.layers 
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd


df = pd.read_csv('StressLevelDataset.csv')

df.columns.tolist()
df["stress_level"].unique().tolist()

X = df.drop("stress_level", axis=1)
Y = df["stress_level"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)

scaler = StandardScaler() 
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

annModel = keras.models.Sequential([
    keras.layers.Dense(32, activation='relu'), #32 and 16 should keep the result consistant
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(3, activation='softmax') #3 for number of opions 0-5 
])


annModel.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = annModel.fit(X_train, Y_train, batch_size=32, epochs=100, validation_data=(X_test, Y_test))

loss, accuracy = annModel.evaluate(X_test, Y_test)
print(f"\nFinal Test Accuracy: {accuracy:.3f}")
print(f"\nFinal Test Loss: {loss:.3f}")

```
