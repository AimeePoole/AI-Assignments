```python
import csv

my_list = []
HSgrad = 0
LessHS = 0
Somecollege = 0
Bachelors = 0
masters = 0

def edu_to_category(edu):
    if edu in ["Preschool", "1st-4th", "5th-6th", "7th-8th", "9th", "10th", "11th", "12th", "Less-HS"]:
        return "Less-HS"
    
    if edu in "HS-grad":
        return "HS-grad"

    #education levels 10-12 more that HS-grad but less that a Bachelors
    if edu in ["Some-college", "Assoc-voc", "Assoc-acdm"]:
        return "Some-college"
    
    if edu in "Bachelors":
        return "Bachelors"

    if edu in ["Masters", "Doctorate", "Prof-school"]:
        return "Masters+"

    return "Other"

with open('adult.data', 'r') as file:
    reader = csv.reader(file)

    for row in reader:
        if len(row) < 15:
            continue

        age = int(row[0].strip())
        workclass = row[1].strip()
        actual_education = row[3].strip()  # This is what we compare against!
        occupation = row[6].strip()
        sex = row[9].strip()
        hours_per_week = int(row[12].strip())
        income = row[14].strip()
        marital_status = row[5].strip()
        capital_gain = int(row[10].strip())
        capital_loss = int(row[11].strip())
        native_country = row[13].strip()

        
        
        value = "<=50K"
        if income == value:
            if edu_to_category("Preschool") == edu_to_category(actual_education):
                LessHS = LessHS + 1

            if edu_to_category("HS-grad") == edu_to_category(actual_education):
                HSgrad = HSgrad + 1

            if edu_to_category("Some-college") == edu_to_category(actual_education):
                Somecollege = Somecollege + 1

            if edu_to_category("Bachelors") == edu_to_category(actual_education):
                Bachelors = Bachelors + 1
            
            my_list.append(edu_to_category(actual_education))
            if edu_to_category("Masters") == edu_to_category(actual_education):
                masters = masters + 1


#for item in my_list[:10]:
#    print(item)
print(f"{value}")
print (f"less: {LessHS}")
print (f"HSgrad: {HSgrad}")
print (f"Somecollege: {Somecollege}")
print (f"Bachelors: {Bachelors}")
print (f"masters: {masters}")
```
